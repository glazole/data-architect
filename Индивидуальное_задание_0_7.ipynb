С помощью языка Python получите данные о текущих курсах валют ЦБ РФ (https://www.cbr-xml-daily.ru/daily_json.js). Отсортируйте список валют в обратном алфавитном порядке (ключ "CharCode"), введите значение CharCode из такого списка для индекса 19:

```python
import requests
```

```python
url = 'https://www.cbr-xml-daily.ru/daily_json.js'
data_j = requests.get(url)
data_json = data_j.json()
print(type(data_json))
print(data_json)
```

Выведем полученный результат в отформатированном представлении с помощью [pprint](https://docs.python.org/3/library/pprint.html):

```python
import pprint
print(pprint.pformat(data_json))
```

```python
data = []
# перебираем
for valute in data_json['Valute']:
    # извлекаем данные
    charcode = data_json['Valute'][valute]['CharCode']
    data.append(charcode)
# сортируем
data1 = sorted(data,reverse=True)
print(data1[19])
```
