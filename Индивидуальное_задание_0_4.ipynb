1. Получите выборку из набора данных  на основании следующих критериев:

Все строки, где  0 и  [79.8125, 81.5234375]

И все строки, где  1 и  [36.4140625, 41.8828125]

```python
link1 = 'https://dc-edu.itmo.ru/asset-v1:ITMOUniversity+DATSC+summer_2022_1+type@asset+block@pulsar_stars_new.csv'
```

```python
df = pd.read_csv(link1)
df
```

```python
df.info()
```

```python
df_new = df[((df['MIP'] <= 81.5234375) & (df['TG'] == 0) & (df['MIP'] >= 79.8125)) | ((df['TG'] == 1 ) & (df['MIP'] >= 36.4140625) & (df['MIP'] <= 41.8828125)) ]
df_new
```

```python
round(df_new['MIP'].mean(),3)
```
