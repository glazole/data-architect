```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/sample_data/Students.xlsx',index_col='ID контрольной/ имя ученика')
df
```

```python
column_of_interest = df.columns[df.isnull().any()][0]
print(column_of_interest)
```

```python
index_of_interest = df[df.isnull().any(axis=1)].index[0]
print(index_of_interest)
```

```python
initial_col = df.columns.to_list()
initial_col.remove(column_of_interest)
print(initial_col)
```

```python
initial_col_expert = initial_col[:4]
print(initial_col_expert)
```

```python
df = df.fillna(0)  # заполняем пропущенные значения нулями
```

```python
df.dtypes
```

```python
df[column_of_interest] = df[column_of_interest].astype(int) # меняем тип колонки на int
```

```python
df['Расстояние'] = 0 # добавляем новую колонку
```

```python
for index, row in df.iterrows():
    sum = 0
    for col in initial_col_expert:
      sum = sum + abs(row[col] - df.loc[index_of_interest,col])
    df.loc[index,'Расстояние'] = sum
```

```python
df
```

```python
df_sorted = df.sort_values('Расстояние') # сортируем по колонке Расстояние
df_sorted
```

```python
df_sorted1 = df_sorted.iloc[1:6, :5] 
df_sorted1
```
