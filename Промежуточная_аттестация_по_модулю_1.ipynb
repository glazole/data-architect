Не все оказалось идеально с данными. Некоторые значения оказались пропущены, некоторые явно содержали экстремальные выбросы. Юрию пришлось обработать исходные данные следующим образом:

Если в описании газонокосилки встречались слова "узкая"/"узкие", то значение восстанавливалось как медиана по существующим в исходном датасете значениям в диапазоне от 30 до 35.


1. Параметр "Ширина скашиваемой полосы" содержал много пропущенных значений. Юрий придумал следующий алгоритм по их восстановлению:

Если в описании газонокосилки встречались слова "узкая"/"узкие", то значение восстанавливалось как медиана по существующим в исходном датасете значениям в диапазоне от 30 до 35.

```python
m_30_35 = df[(df["WIDTH"] >= 30) & (df["WIDTH"] <= 35)]['WIDTH'].median() #находим медиану
m_30_35
```

Если встречались слова "средняя ширина"/"ширина средняя", то значение восстанавливалось как округленное до целого среднее арифметическое по существующим значениям в диапазоне от 36 до 40.
Введите округленное до целого среднее арифметическое по существующим в исходном датасете значениям в диапазоне от 36 до 40.

```python
m_36_40 = df[(df["WIDTH"] >= 36) & (df["WIDTH"] <= 40)]['WIDTH'].median() #находим медиану
m_36_40
```

Если встречались слова ("широкая" и "полоса")/("широкие" и "полосы"), то значение восстанавливалось как медиана по существующим значениям в диапазоне от 41 до 45.
Введите медиану по существующим в исходном датасете значениям в диапазоне от 41 до 45.

```python
m_41_45 = df[(df["WIDTH"] >= 41) & (df["WIDTH"] <= 45)]['WIDTH'].median() #находим медиану
m_41_45
```

Востановим пропущенные значения

1 вариант

```python
df.loc[(df['DESCRIPTION'].str.contains(r'узкая|узкие|узкую|узкой', case=False)) & (df['WIDTH'].isnull()), 'WIDTH'] = m_30_35
df.head(20)
```

 2 вариант

```python
pattern = re.compile(r'узкая|узкие|узкую|узкой', re.IGNORECASE)
df['WIDTH'] =df.apply(lambda x: m_30_35 if re.search(pattern, x['DESCRIPTION']) and pd.isna(x['WIDTH']) else x['WIDTH'],axis = 1)
df.head(20)
```

```python
pattern = re.compile(r'средняя ширина|ширина средняя', re.IGNORECASE)
df['WIDTH'] =df.apply(lambda x: m_36_40 if re.search(pattern, x['DESCRIPTION']) and pd.isna(x['WIDTH']) else x['WIDTH'],axis = 1)
pattern = re.compile(r'широк(?:ая|ие)\s+полос(?:а|ы)', re.IGNORECASE)
df['WIDTH'] =df.apply(lambda x: m_41_45 if re.search(pattern, x['DESCRIPTION']) and pd.isna(x['WIDTH']) else x['WIDTH'],axis = 1)
```

```python
df.head(20)
```

Cреднее арифметическое для параметра "Ширина скашиваемой полосы" после восстановления пропущенных значений:

```python
print(round(df["WIDTH"].mean(),2))
```

2. Цены на газонокосилки имели очень большой разброс и Юрий решил избавиться от вариантов, стоимость которых попадала под определение "экстремальных выбросов".

Находим значения квартилей

```python
Q1 = df['PRICE'].quantile(0.25)
Q2 = df['PRICE'].quantile(0.5)
Q3 = df['PRICE'].quantile(0.75)
IQR = Q3-Q1
print(Q1,Q2,Q3,IQR)
print(IQR*3)
```

```python
print(Q1 - 3 * IQR)
print(Q3 + 3 * IQR)
```

Удаляем экстремальные выбросы

```python
df = df[(df['PRICE'] > Q1 - 3 * IQR) & (df['PRICE'] < Q3 + 3 * IQR)]
df
```

```python
df['PRICE'].sum()
```

Исключаем записи, в которых все еще оставались пропущенные значения.

```python
df = df.dropna() 
df
```

4. Параметры газонокосилок имеют разную размерность, так что Юрий решил экспоненциально нормировать значения числовых параметров.

```python
df_n = df.drop(columns=['DESCRIPTION'])
df_n = 1 - np.exp(1 - df_n/df_n.min())
df_n
```

Введите нормированные значения параметров газнокосилки Worx EDL764:

```python
print(round(df_n.loc['Worx EDL764']['WIDTH'],2))
print(round(df_n.loc['Worx EDL764']['PRICE'],2))
print(round(df_n.loc['Worx EDL764']['POWER'],2))
print(round(df_n.loc['Worx EDL764']['GRASS'],2))
print(round(df_n.loc['Worx EDL764']['AREA'],2))
```

 

5.   И наконец, Юрий составил целевую функцию, на основе которой определил 3 наиболее подходящие (на его взгляд) газонокосилки. При определении целевой функции он использовал следующие слагаемые:

Нормированная ширина скашиваемой полосы с коэффициентом 1
Стоимость как слагаемое вида (1 - нормированная стоимость) с коэффициентом 8
Нормированную мощность с коэффициентом 8
Нормированный объём травосборника с коэффициентом 5
Нормированную площадь скашиваемой поверхности с коэффициентом 5

```python
df_n['SUM'] = df_n['WIDTH'] + (1 - df_n['PRICE']) * 8 + df_n['POWER'] * 8 + df_n['GRASS'] * 5 + df_n['AREA'] * 5
df_n.sort_values(by = ['SUM'], ascending = False)
```
