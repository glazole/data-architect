```python
import pandas as pd
import numpy as np
import datetime
```

```python
data = pd.read_csv('/content/sample_data/Индивидуальное задание_1_2.csv')
data
```

```python
data.info()
```

Преобразуем дату в удобный для дальнейшего формат, склеив дату и время

```python
data['DATE'] = data['<DATE>'] + ' ' + data['<TIME>'] 
data.head()
```

Переведем строковые данные, показывающие время, в специальный объект в Python.

```python
print(type(data['DATE'][1]))
```

```python
data['DATE'] = pd.to_datetime(data['DATE'])
data.head()
```

Теперь полная дата будет для нас индеком объектов

```python
data = data.set_index('DATE')
data
```

Добавим колонку <code>'PRICE'</code>




```python
data['PRICE'] = data[['<OPEN>','<CLOSE>', '<LOW>', '<HIGH>']].mean(axis=1)
```

```python
data
```

Сформируем интересующую нас выборку для пункта 1

```python
data_filtered = data.loc['2020-10-26 12:12:00']
```

```python
data_filtered
```

Добавим колонку <code>'TOTAL'</code>

```python
data['TOTAL'] = data['<VOL>'] * data['PRICE']
print(data)
```

Сформируем интересующую нас выборку для пункта 2

```python
data_filtered = data.loc['2020-12-10 16:28:00']
data_filtered['TOTAL']
```

Сформируем интересующую нас выборку для пункта 3

```python
data_filtered = data.loc['2020-12-17']
print(data_filtered)
round(data_filtered['TOTAL'].sum())
```

Сформируем интересующую нас выборку для пункта 4 (количество минутных интервалов, когда цена открытия была строго больше цены закрытия).

```python
data_filtered = data.loc['2020-03-09']
data_filtered[data_filtered['<OPEN>'] > data_filtered['<CLOSE>']].shape[0] 
```
