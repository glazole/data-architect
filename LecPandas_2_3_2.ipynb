# Анализ данных в Pandas

---

Импортируем библиотеки и загрузим данные:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/sample_data/avito_flats.csv', sep=';', index_col=0)
print(df.columns)
```

```python
df['Цена'] = df['Цена'].str.replace(' ', '') #Преобразуем такие строки с ценой, как '9 616 960' к '9616960', а далее все численные данные к типу  np.float64
df = df.astype({'Цена': np.float64})
```

---

## 2 | Визуализация

---

Для того чтобы использовать круговую диаграмму, надо использовать метод $\mathsf{plot.pie}$ после вызова знакомого нам метода $\mathsf{value\_counts}$:

```python
df['Материал_стен'].value_counts().plot.pie(autopct="%.1f")
```

---

Помимо круговых диаграмм есть столбчатые, за отрисовку которых есть функция $\mathsf{plot.bar}$:

```python
df['Материал_стен'].value_counts().plot.bar()
```

---

Предыдущие диаграмы используются только для номинальных признаков, а вот для количественных можно использовать гистограммы, которые можно получить, вызвав метод $\mathsf{plot.hist}$:

```python
df['Цена'].plot.hist()
```

Для более детального отображения можно регулировать параметр $\mathsf{bins}$, отвечающий за количество столбиков:

```python
df['Цена'].plot.hist(bins=5)
```

```python
df['Цена'].plot.hist(bins=50)
```

```python
df['Цена'].plot.hist(bins=20000)
```

Для таких данных, которые похожи на экспоненциальное распределение, бывает иногда полезно применить функцию логарифма перед самим построением гистограммы:

```python
np.log10(df['Цена']).plot.hist(bins=25)
```
